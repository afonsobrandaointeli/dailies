```python
from pymongo import MongoClient
from faker import Faker
import random
from datetime import datetime, timedelta

# Conectando ao MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["daily_db"]
email_collection = db["emails"]
response_collection = db["responses"]

# Instanciando o Faker
fake = Faker()

# Lista de e-mails de teste para 35 estudantes
emails = [f"estudante{i}@inteli.edu.br" for i in range(1, 36)]

# Grupos e turma
grupos = ["Grupo A", "Grupo B", "Grupo C", "Grupo D", "Grupo E"]
turma = "t014"

# Função para gerar respostas falsas
def gerar_resposta_falsa(email):
    data = fake.date_between(start_date='-30d', end_date='today')
    tarefa_realizada = fake.sentence()
    progresso = random.choice(["Concluído", "Em Progresso", "Obstáculos Encontrados"])
    descricao_obstaculos = fake.sentence() if progresso == "Obstáculos Encontrados" else ""
    proximas_etapas = fake.sentence()
    comentarios_adicionais = fake.sentence()

    return {
        "email": email,
        "data": data.isoformat(),
        "tarefa_realizada": tarefa_realizada,
        "progresso": progresso,
        "descricao_obstaculos": descricao_obstaculos,
        "proximas_etapas": proximas_etapas,
        "comentarios_adicionais": comentarios_adicionais
    }

# Função para gerar dados de e-mails com grupo e turma
def gerar_dados_email(email, grupo, turma):
    return {
        "email": email,
        "grupo": grupo,
        "turma": turma
    }

# Inserindo e-mails na coleção de e-mails
email_collection.delete_many({})  # Limpar coleção
emails_data = [gerar_dados_email(email, random.choice(grupos), turma) for email in emails]
email_collection.insert_many(emails_data)

# Gerando e inserindo dados falsos na coleção de respostas
response_collection.delete_many({})  # Limpar coleção
responses = []

for email in emails:
    for _ in range(30):  # Gerar 30 respostas por e-mail
        responses.append(gerar_resposta_falsa(email))

response_collection.insert_many(responses)

# Verificação dos dados inseridos
print(f"Total de e-mails inseridos: {email_collection.count_documents({})}")
print(f"Total de respostas inseridas: {response_collection.count_documents({})}")
```

```python
from pymongo import MongoClient
import pandas as pd
from IPython.display import display

# Conectando ao MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["daily_db"]

# Função para mostrar todos os dados de todas as coleções
def mostrar_dados_mongo():
    colecoes = db.list_collection_names()
    
    for colecao in colecoes:
        print(f"--- Dados da Coleção: {colecao} ---")
        dados = list(db[colecao].find())
        if dados:
            df = pd.DataFrame(dados)
            display(df)
        else:
            print("A coleção está vazia.")
        print("\n")

# Executar a função para mostrar os dados
mostrar_dados_mongo()
```
