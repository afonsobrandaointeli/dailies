```python
import psycopg2
from psycopg2 import sql
from dotenv import load_dotenv
import os

# Carregar variáveis de ambiente do arquivo .env
load_dotenv()

# Conectar ao banco de dados PostgreSQL
conn = psycopg2.connect(
    host=os.getenv("DB_HOST"),
    dbname=os.getenv("DB_NAME"),
    user=os.getenv("DB_USER"),
    password=os.getenv("DB_PASSWORD")
)
cursor = conn.cursor()

# Comandos SQL para criar as tabelas
create_emails_table = """
CREATE TABLE IF NOT EXISTS emails (
    email VARCHAR(255) PRIMARY KEY
);
"""

create_responses_table = """
CREATE TABLE IF NOT EXISTS responses (
    id SERIAL PRIMARY KEY,
    email VARCHAR(255) REFERENCES emails(email),
    data DATE NOT NULL,
    tarefa_realizada TEXT NOT NULL,
    progresso VARCHAR(50) NOT NULL,
    descricao_obstaculos TEXT,
    proximas_etapas TEXT,
    comentarios_adicionais TEXT
);
"""

# Executar os comandos SQL
cursor.execute(create_emails_table)
cursor.execute(create_responses_table)

# Confirmar as mudanças e fechar a conexão
conn.commit()
cursor.close()
conn.close()

print("Tabelas criadas com sucesso!")
```
